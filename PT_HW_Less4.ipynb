 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
    Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
    Если да, найдите ее.

```python
Dx=0.2
a=0.5
b=math.sqrt(12*Dx)+a
print('Правая граница: ',b)
print('Среднее значение:', (b+a)/2)
```

 4. Рост взрослого населения города X имеет нормальное распределение.
    Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
    Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
  

 а). больше 182 см. То есть нужно найти x>182, что соответствует 1-P(для X<=182). Находим какому значению сигма соответствует значение 182. Находим X-U(Мю)/среднее квадратичное, то есть (182-174)/8=1 (то есть сигма = +1). Согласно таблице накопленного нормального распределения сигма +1 соответсвует значение нормального распределения 0,8413. Соответственно 1-0,8413=0,1587 или 15,87%

 б). больше 190 см. Ищем x>190. Находим 1-P(для X<=190), т.е. сигма = +2, значение нормального распределения 0,9713. То есть 1-0,9713 = 0,0287 или 2,87%

 в). от 166 см до 190 см. Ищем 166<x<190. 
x>166 = 1-P(для x<=166; сигма =-1) = 1-0,1587=0,8413; 
x<190, сигма = 2, 0,0228. То есть получаем, что x>166 имеет распределение 0,8413, x>=190 имеет распределение 0,0228. Из объема x>166 нужно вычесть объем X>190, то есть 0,8413-0,0228= 0,8185 (81,85%) 

 г). от 166 см до 182 см. значение 166 есть сигма =-1, значение 182 есть сигма =+1, то есть вероятность между -1 сигма и +1 сигма составляет 68,26%. 
 Для проверки можно выполнить вычисления:
 Ищем веротяность для 166<x<182.
 Вероятность x>166 есть 1-P(x<=166), т.е. 1-0,1587=0,8413
 Вероятность x<182 есть 0,8413
 Таким образом P(x<=166)=0,1587, P(x>=182)=1-0,8413=0.1587
 166<x<182 = 1-(x<166)+(x>182)=1-0,1587+0,1587=0,6826(68,26%)
 

 д). от 158 см до 190 см. Аналогичное решение при условии, что 158 - сигма(-2), 190 - сигма(+2), то есть 95,4%

 x<150; сигма =-3; P=0,0014
 x>190; 1-P(X<=190; сигма=2) получаем 1-0,9772=0,0228
 0,0014+0,0228=2,42

5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
    математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?
    Решение: так как D(x)=25, следовательно сигма = SQRT(25)=5. Далее находим (190-178)/5= 2,4 сигмы.
    Ответ: 2,4 сигмы.
