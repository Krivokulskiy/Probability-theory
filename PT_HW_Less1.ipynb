1) Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

Рассуждения: В колоде 52 карты, исходим из того что масти распределены одинаково, таким образом получаем, что количество карт для каждой масти равно 52/4=13. То есть 13 карт из 52 имеют масть крести. Сочетание 4 из 13 есть число благоприятствующих искомому событий для карт масти крести. Сочетания 4 из 52 (общему числу карт) общее число всех возможных комбинаций карт.

```python
from math import factorial
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
C1=combinations(13,4)
print('количество сочетаний 4 карт мас ти крести: ', C1)
C2=combinations(52,4)
print ('количесто всех сочитаний 4 карт из 52 карт:', C2)
print ('Искомая вероятность выпадения 4 карт масти крести: ', C1/C2, 'процентов')
```

```python
C1=combinations(4,1)
print('количество сочетаний 4 карт из которых 1 является тузом: ', C1)
C2=combinations(48,3)
print('количество сочитаний 48 карт по 4, исключая тузы: ', C2)
print('количество сочетаний для 52 карт по 4 карты из которых 1 является тузом: ', C1*C2)

C3=combinations(52,4)
print ('количесто всех сочитаний 4 карт из 52 карт:', C3)
print ('Искомая вероятность выпадения 4 карт масти крести: ', C1+C2/C3*100, 'процентов')
```
